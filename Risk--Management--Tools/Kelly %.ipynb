# Kelly % = W/A - (1 - W)/B
## W = win--Rate (%)
## A  = Avg--Losing--Trade (%)
## B  = Avg--Winning--Trade (%)


### Exemple :

```python
W = 0.5  # Win Rate [%]   
A = 0.5  # Avg Losing Trade [%]
B = 0.51 # Avg Winning Trade [%] 
print('Le montant de ton capital total disponible sur chaque trade: ',round((W/A) - ((1-W)/B),3)*100,'%')
```

### Risque 1:1 evolution selon le winrate :

```python
liste = []
liste_2 = []
B = 1
A = 1
for i in range(1,100):
    
    W = i/100
    liste.append((W/A) - ((1-W)/B))
    liste_2.append(i)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'Win--Rate':liste_2,'% to risk':liste})
```

```python
df.index = df['Win--Rate']
df['0'] = 0
```

```python
df['% to risk'].plot(figsize=(10,6))
plt.grid()
plt.plot(df.index,df['0'],c='red',lw=3)
```
